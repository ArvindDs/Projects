```python
# import the library
import pandas as pd            #for reading & storing data, pre-processing
```

```python
#read the dataframe
# checkout the first five rows
df=pd.read_csv('/content/drive/MyDrive/faa_ai_prelim.csv')
df.head()
```

```python
# create a new dataframe with only required coumns.
df_new=df[['UPDATED',	'ENTRY_DATE',	'EVENT_LCL_DATE',	'EVENT_LCL_TIME',	'LOC_CITY_NAME',	'LOC_STATE_NAME','LOC_CNTRY_NAME','ACFT_MAKE_NAME','FATAL_FLAG']]
```

```python
# view the new data set
df_new.head()
```

```python
# view the type of the object
type(df_new.head)
```

```python
# replace NaN in LOC_CNTRY_NAME with No
df_new['LOC_CNTRY_NAME'].fillna(value='NO',inplace= True)
df_new['FATAL_FLAG'].fillna(value='NO',inplace= True)
```

```python
# view first five obserations
df_new.head()
```

```python
df_new.shape
```

```python
# drop values where aircraft make name ACFT_MAKE_NAME is not avilable
df_new_final=df_new.dropna(subset=['ACFT_MAKE_NAME'])
```

```python
df_new_final.shape
```

```python
# group by aircraft name
group_aircraft_name=df_new_final.groupby(['ACFT_MAKE_NAME'])
```

```python
group_aircraft_name.head()
```

```python
# view the size of the name.       mean how ment times a make name appeared in the dataset
group_aircraft_name.size()
```

```python
# group by FATAL_FLAG accident name
accident_fatal=df_new_final.groupby(['FATAL_FLAG'])
```

```python
accident_fatal.size()
```

```python
# select the accident with fatality yes
fetal_accident=accident_fatal.get_group('Yes')
```

```python
# view the accident with fatality
fetal_accident
```
