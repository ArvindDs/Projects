```python
# import required libraries
from bs4 import BeautifulSoup
import requests
```

```python
# url for web scraping
url= ('http://www.simplilearn.com')
```

```python
# acess the url with request object
result=requests.get(url)
# scrape the web page content
webpage=result.content
#create a soup object for parsing web page using html paraser
sl_soup=BeautifulSoup(webpage,'html.parser')
```

```python
# close the  result object
result.close()
```

```python
# View the content  of  the soup object
sl_soup.contents
```

```python
# use prettify method to view well formatted output(html tags)
print(sl_soup.prettify())
```

```python
# view the ktitle of the web page
sl_soup.title
```

```python
# find all the llinks present on the web page
for href in sl_soup.findAll('a',href=True):
  print(href['href'])
  
```
