```python
df=pd.read_csv('/content/drive/MyDrive/auto_data.csv')
df.head()
```

```python
#write a user defined function for origin of the cars. here the origin is defined by 1,2,3 so define it by USA, Euorpe and Asia respectively.
def define(num):
  if num == 1:
    return 'USA'
  elif num== 2:
    return 'Europe'
  else:
    return 'Asia'

```

```python
df.head()
```

```python
# use apply function for applying these codes in data frame
df['origin']=  df['origin'].apply(define)
```

```python
# draw a pair plot using sns for mpg,weight and with  hue origin ,set the size to 4
# hue is variable in a dstaset to map plot aspects to different colors
import seaborn as sns
sns.pairplot(df[['mpg','weight','origin']],hue='origin',height=4)
```
